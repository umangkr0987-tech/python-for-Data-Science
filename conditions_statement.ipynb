```python
# Conditional Statement----------------------------- (If-else)
           # if - excutes inner code , if the condition is true
           # if-else - excutes => if is True , else for false
           # if-elif-else  => checks multiple condition in sequence

# If syntax------ if condition: code
a=13
if a>10:
    print("I will do task A")
else:
    print("I will do task B")

#PRACTIC SERIES-----------------------------
money = int(input("Please provide me the money in 10/20 rupees:"))
if money==10:
    print("Yeah! I can buy Chocolate flavour ice-cream")
else:
    print("Hooray! I can buy Vanilla Flavour ice-cream")
    
# For Multiple Condition-----------  use if-elif-else

money= int(input("Please provide the money in the form of thousand and between 60000 & 1000000:"))

if money>60000 and money<80000:
    print(" Suggestion:Buy a Bullet motorbike")
elif money<500000:
    print("Suggestion: Buy a Suzuki Car")
elif money<1000000:
    print("Suggestion: Buy a Creata Car")
else:
    print("Suggestion: Please Start saving the money")

# PRACTICE SERIES--------------------------------------------------------------------------------
# Find greatest no. between two input numbers
no1= int(input("Enter the first no.:"))
no2= int(input("Enter the second no.:"))

if no1>no2:
    print(f"Greatest number is :{no1}")
elif no2>no1:
    print(f"Greatest number is :{no2}")
else:
    print("Both No. are same")

# Greeting message by the gender------
gender=input("Please tell your gender as character (M or F):")

if gender=="M" or gender=="m":
    print("Good Mornning Sir! Have a nice day...")
elif gender=="F" or gender=="f":
    print("Good Morning Mam! Have a beautiful day...")
else:
    print("Please! give the correct indentified gender dear")

#Check ODD or EVEN no:---------------
num= int(input("Enter the number:"))

if num%2==0:
    print(f"{num} is a even number")
elif num%2!=0:
    print(f"{num} is a odd number")
else:
    print("Please! Give the input as a number")

# Check the user is a valid voter or not----------------
name=input("Enter your name:")
Age= int(input("Enter your age:"))

if Age>=18:
    print(f"Hello {name}, you are a valid voter")
elif Age<18:
    print(f"Hello {name}, Sorry! you are not a valid voter")   

# Check year is Leap year or not------------------------------

year= int(input("Tell the year:"))

if year%400==0 :
    print(f"{year} is a leap year")
elif year%4==0 and year%100!=0:
    print(f"{year} is  a leap year")
else:
    print(f"{year} is not a leap year")

# Practice If-elif Ladder question-------------------------------

temp=int(input("Enter the temperature in the form of celcius:"))

if temp<0:
    print("Freezing Cold !")
elif temp>=0 and temp<10:
    print("Very cold !")
elif temp>=10 and temp<20:
    print("Cold !")
elif temp>=20 and temp<30:
    print("Pleasent !")
elif temp>=30 and temp<40:
    print("Hot !")
else:
    print("Very Hot !")
```
