```python
#Functions-----------------------------------------
              #Two types of function----- in-built function, user defined function

# Example of in-built function => print(),range(),len() etc.

#User defined function uo hota hai , jo humlog create krte hai-------------
                  #Syntax-- def func_name(): code

def hello():
    print("Good morning dear!")

hello()   # function call => function create krne ke baad usko call krna padta hai, tb function excute hota hai  
hello()

#Function parameters and arguments-------------------
                            #Parameters => jo variables/value , function create ke samay pass hota hai
                            #Arguments => Jo value/variables, function call ke samay pass hota hai

def sum(a,b):                                  # a,b is a Parameters
    print(f"The total sum is :{a+b}")

no1=int(input("Enter the first number:"))
no2=int(input("Enter the second number:"))
sum(no1,no2)                                    # no1,no2 is a arguments

#Types of parameters/arguments------------------------------------------------
       # Three types--Positional parameters/arguments, Default parameters/arguments, keyword arguments

#Positional Arguments---------isme argument ko pass kiya gya value, parameter ke variables me same order me assign hota hai, (jitna argument pass pass hoga ,utna parameter bhi hona chaiye)

def mul(a,b):                                     # 2 Parameters
    print(f"multiply={a*b}")

num=int(input("Enter the first number:"))
numm=int(input("Enter the second number:"))

mul(num,numm)  # 2 arguments  , num is assinged to a, numm is assigned to b

#Default Parameters------------ isme parameter ka variable me pahle se hi value de diya jata hai, taki agar argument me value pass na ho, to default ye value use ho jaye

def greet(name="Guest"):         # default parameter (name="Guest")
    print(f"Good Morning {name}")

greet()    
greet("umang")

#Keyword Arguments------ agar order maintain ni krna ho Positional Argumemts jaisa, to jo variable parameter me pass kiye hote hai, wahi variable ke sath value ko function call me pass kr diya jata hai ,isse order maintain krne ka jarurat ni padta

def intro(name,age):
    print(f"Hey {name} ! Your age is {age}")

intro(age=21,name="Umang")     # Keyword Argument (age=21, name="Umang")

#Practice Series---------------------------------------------------

#Create a function to check input string is palindrome or not?

def palindrome(str1):
    b=""
    for i in range(len(str1)-1,-1,-1):
        b+=str1[i]

    if b==str1:
        print(f"{str1} is a palindrome string")
    else:
        print(f"{str1} is not a palindrome string")



inp= input("Enter the string:")
palindrome(inp)

# Return keyword------function definition me return keyword ka use,function se output/result ko dobara function call ke pass bhejne me hota hai 

def sub(a,b):
    return a-b

c=sub(5,8)     # function se output sub(5,8) function call ke pass aa gya hai, jisko variable c me store kr rhe
print(c)
```
