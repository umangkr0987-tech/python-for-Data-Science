```python
#Tuple Data Structure-----------------------------------------------------------------------------
                         # Syntax: variable_name= (value1,value2,value3,......)

#Property of Tuple-------------------- 1) Immutable :Tuple create krne ke baad uska value ko change ni kiya ja sakta hai..
                                    #  2) Duplicate: Tuple duplicate value allow krta hai
                                    #  3) Ordered : Tuple me value indexed wise store rhte hia , jisse index ke through value ko access kiya ja sakta hai..
                                    #  4) Hetrogenous : Tuple me multiple datatype ka value , function etc. sab ek hi Tuple me store ho sakta hai...

a=(4,5,6,3,9,1,True,5.5,6.458,print())    
#a[1]=23                   # no value change, once created
print(a)
print(type(a))

# Tuple me "Indexing" aur "Slice" same hota hai string jaisa....

b=(67,54,34,32.12,45,98,56.8)

print(b[2])   # Index ke through access value
print(b[-1])  # Negative indexing ke help se bhi access kr sakte hai
print(b[1:5:1])  # Slice krna ( Tuple ka koe part access krna)

# Tuple me sara value pr "Traversing" bhi same string traversing jaisa hi "Loop" ke help se hota hai...
c= [12,23,34,45,56,67,78,"hello"]

#Using index Traversing-----------
for i in range(len(c)):
    print(c[i])

# Using direct values traversing------
for i in c:
    print(i)

#Tuple Method--------------------------------------------------------------------
                  #Method ko access krne ka Syntax---- Tuple_variable_name.method_name()

#Only 2 Methods----------------

# 1) Count(value) => ye jo value pass hota hai, uo kitna baar aaya hai Tuple me ,uo count kr deta hai, jisko humlog ek variable me store kr sakte hai
z=[4,5,6,7,8,4,5,4,5,4,5]
coun= z.count(5)
print(coun)  

# 2) index(value) => isse jo value pass hoga , uss value ka Tuple me index no. pta kr sakte hai,lakin only first occurance (jo first baar Tuple me aaya ho...),jisko ek varible me store krkr baad me use kr sakte hai
w=[5,2,2,3,8,9]
pos=w.index(2)     # 2 ka index no. Tuple me pta find krna
print(pos)

# Concept of Tuple Unpacking--------------------------------

#Tuple Unpacking ka mtlb => agr koe tuple me x value hai, aur usko x no.of variable me store kr rhe hai, to har variable me single single value assign ho jayega.

p,q,r=("umang",6,6.25)  # Tuple unpacking => vaiable "p" me (umang) assign ho gya, variable "q" me (6) assign ho gya aur variable "r" me (6.25) assign ho gya

print(p)
print(q)
print(r)

# Question-------------
m=(6)   # what datatype? (Tuple/int)

print(type(m))  # "int datatype"  => Q ki tuple me ek hi value hai , aur usko ek hi variable me store kr rhe , isiliye ye indirectly " Tuple Unpacking" ho gya

n=(3,)
print(type(n))  # "tuple datatype" => Q ki comma ke wajah se tuple ke andar 2 value lag rha ,aur ek hi variable me store hone ke karan  "Tuple Unpacking" ni ho paya

# Practice Series--------------------------------------------------------------------------------------

# Convert Tuple into List
s=(4,6,8,2,9)
b=[]

for i in range(len(s)):
    b.append(s[i])

print(b)   

```
