```python
#Set Data Structure-----------------------------------------------------------------------------
                         # Syntax: variable_name= {value1,value2,value3,......}

#Property of Set--------------------- 1) Mutable :Set create krne ke baad uska value ko change kiya ja sakta hai..
                                    #  2) No Duplicate: Set duplicate value allow ni krta hai, (every element must be unique)
                                    #  3) Unordered : Set me value indexed wise store ni hota hai, jiske karan index ke through value ko access ni kiya ja sakta hai..
                                    #  4) Semi-Hetrogenous : Set me kuch datatype ka value sath me store ho sakta hai..(String,Boolean, number,tuples)

# Agar set me pahle se koe duplicate element present hai, to bhi output ke samay ,Set duplicate element ko ignore kr deta hai

# Set me har element ,"hash value" ke through store hota hai, aur "hash value" koe bhi element ka fix ni hota hai (jitna time run krenge set ko , utna baar hash value har element ka alag alag aayega)

a={1,5,"aadarsh" ,7,3,"umang",9,(3,5,6,7)}
print(a)
print(type(a))

#Set ke elements pr traversing ke liye index ka use ni hota hai, loop ke help se direct element/value ke base pr travering kr sakte hai...

for i in a:
    print(i)

#Set Method--------------------------------------------------------------------
                  #Method ko access krne ka Syntax---- Set_variable_name.method_name()
    
# 1) add(value)  => ye value ko Set ke first me add kr deta hai
l= {2,4,6,7,9,4,7}
l.add(17)    
l.add(True)
print(l)  

# 2) remove(value) => ye jo value pass hota hai, uss value ko Set se remove kr deta hai
n={2,3,4,5,6,7,4,5}
n.remove(4)           
print(n)

# 3) clear() => isse Set ke sara elment ko remove kr sakte hai
r={4,5,6,7,8,9,23}
r.clear()
print(r)

# 4) pop() => isse Set ka koe bhi element randomly remove ho sakta hai, aur jo element remove hoga usko humko ek variable me store krke baad me use kr sakta hai
b={7,1,3,5}
gayab=b.pop()
print(gayab)       # jo element randomly remove hua hoga, uo "gayab" variable me store ho jayega
print(b)

# Other Important Methods-----------------------------------

# Agar koe 2 Set ke bich humlog ko un dono Set ke bich => "Union","Intersection","Set_difference","Symmetric-difference" nikalna hai to---

r={1,2,3,4,5}
s={4,5,6,7,8,9}

# "Union" ke liye => Syntax--- "Set1_variable_name.union(Set2_variable_name)" or "Set1_variable_name|Set2_variable_name" , isko humlog ek variable me store kr sakte hai / ya  Set1 or Set2 me hi change krna ho (union nikal kr), to Set1 or Set2 varibale me bhi store kr sakte hai
t= r.union(s)
print(t)

w= r|s
print(w)

# "Intersection" ke liye => Syntax--- "Set1_variable_name.intersection(Set2_variable_name)" or "Set1_variable_name & Set2_variable_name" , isko humlog ek variable me store kr sakte hai / ya  Set1 or Set2 me hi change krna ho (intersection nikal kr), to Set1 or Set2 varibale me bhi store kr sakte hai
e= r.intersection(s)
print(e)

f= r&s
print(f)

# "Set-Difference" ke liye => Syntax--- "Set1_variable_name.difference(Set2_variable_name)" or "Set1_variable_name - Set2_variable_name" , isko humlog ek variable me store kr sakte hai / ya  Set1 or Set2 me hi change krna ho ( Set-difference nikal kr), to Set1 or Set2 varibale me bhi store kr sakte hai

v= r.difference(s)
print(v)

p= r-s
print(p)

# "Symmetric-difference" => mtlb ki dono Set me common element ko chor ke dono Set ke baki element ko return krna

# "Symmetric-difference" ke liye => Syntax--- "Set1_variable_name.symmetric_difference(Set2_variable_name)" or "Set1_variable_name ^ Set2_variable_name" , isko humlog ek variable me store kr sakte hai / ya  Set1 or Set2 me hi change krna ho (Symmetric_differernce nikal kr), to Set1 or Set2 varibale me bhi store kr sakte hai

j= r.symmetric_difference(s)
print(j)

k= r^s
print(k)


# Set ke elements ko separately access ni kiya ja sakta hai, Q ki elements sara hash value ke through randomly store hote hai
```
