```python
print("Hello, I am Umang and I am learning DATA SCIENCE ")

#Comments in python----------------------------------------------------------------------------------
#Comments are something that are ignored by the python interpreter. We have to use for writing a comment in python
#Multiline comments are not available in python but we can achieve it by using Doc String, but its not comment ,its use to written useless multi lines like paragrapgh , SYNTAX--
""" Multiline """ 

#Variable in Python-----------------------------------------------------------------------------------

# For numbers-- Integer, Float, Complex

abc=1.5
b=-6
umang="data science"
c=12/5   
rm=12/3    # iska answer 4 ni balki, 4.0 aayega jo ki ek float value hai Q ki, jb koe no. p/q ke rup me likhata hai to uska answer float datatype me consider kiya jata hai..
v=65j  # Complex datatype--  iota (i=-1), yaha isko python me "j" se represent kiya jata hai.

print(type(umang))
print(type(b))
print(type(abc))
print(type(c))
print(type(rm))
print(type(v))

# For String

st="1234565"
str2="aadarsh"
str1="@#$%^^%$##"

print(type(st))
print(type(str2))
print(type(str1))

#For Boolean

vn= True
tb= False  # Always first letter capital 

bn="True"

print(type(vn))
print(type(tb))
print(type(bn))
```

```python
# How String work----  Every string has indicate UNICODE (number) . Ex:- A= 65 etc.-------------------------

bc="A"
print(ord(bc))  # character convert into UNICODE through => ord()

ca= 65
print(chr(ca)) # UNICODE convert into character through => chr()

da= 128522
print(chr(da))

#String Indexing => String ke koe character ko aacess krne ke liye string ko indexing kr dena---------------

uma= "Aadarsh"
print(uma[5])  # positive Indexing left -> right jata hai , 0 se start hoke 1,2,3...

print(uma[-2])  # negative indexing right -> left jata hai , -1 se start hokr -2, -3,-4...

print(uma[2],uma[-5])

# String slicing-----pura string ya string paragraph me se koe string part ko acccess krne ka tarika--------

bh = "Sher coder"
print(bh[1:4:1])  # Syntax:-- string_name[start_index : last_index+1 : no. of step going forward]
```

```python
# Type Conversion----------- ek datatype se dusre data type me change krna, [ main 4 datatype: int(), float(), str(), bool()]

av= 65
print(type(av)) # integer type

bk=float(av)  #change datatype => float
print(type(bk))  #float type
print(bk)
print(av)

bn="56"    #String type
ahn=int(bn)  #change datatype => integer

print(type(bn))
print(type(ahn))

gb=98            # integer type
nm=str(gb)       # change datatype => string

print(type(gb))   #interger type
print(type(nm))   #string type

xo= 67
print(bool(xo))   # change bool datatype, but only print case NOT actual
print(type(xo))   # Integer type

mm="XS-AARU"
print(bool(mm))

mmn=False
print(type(mmn))  # BOOl type

bbh=""
print(type(bbh))  # String type
print(bool(bbh))  # change BOOL type, but only print case NOT actual
print(type(bbh))  # String type

# EXplicit Type Conversion , humlog manually krte hai => int(), float(), str() etc. function ke help se
# IMplicit Type Conversion , python automatically kr deta hai , khud se code ke likhne ka tarika ka samjh kr

```
