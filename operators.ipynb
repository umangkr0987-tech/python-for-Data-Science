```python
# Operators----------------------------------------------------

#Airthmethic operator------- mathematical operation perform (+, -, * ,/ ,% ,// ,** )

ak=9
uk=6
print(ak+uk)
print(ak-uk)
print(ak*uk)

lk=20
uk=5
print(lk/uk)   # answer 4 ni aayega balki 4.0 aayega Q ki p/q ke form me rhne pr answer float datatype me aa jata hai

print(lk//uk)  # // =>floor operator, iska kaam bhi division krna hi hai, but ye answer me integer value deta hai (agar decimal me answer aa rha to nearest integer me answer dega)

print(89//11)

print(5**2)  # ** => power operator (a**b => a ka power b hai)
print(5**100)

print(32%5) # % => modulus operator (iska kaam hai divide krke , remainder ko answer ke rup me dena)

# Python BODMAS Rule follow krta hai, mathemathical operation ko solve krte time
```

```python
# Assignment Operator--------------------------------------- variable  me value assign krna aur manipulate krne wala operator
# Equal operator--

a=20 # variable a me 20 assign ho gya

# variable me value reassign krna-----

a=a+40   # a+=40 (shortcut), sara airthmethic operator use ho sakta hai...( a-=, a*=,a/= etc)
print(a+80)

# Comparision Operator-------------------------( always return bollean value -true/false )
                 # six operator - (>,<,>=,<=,==,!=)

a=2.0
b=2
print(a==b) # python me FLOAT-INTEGER datatype value bhi compare ho jata hai
print(a>=b)
print(a!=b)

c=87
d=54
print(c!=d)
print(c>d)
print(89>89.1)

print("A">"B")  # string bhi compare ho sakta hai, but Unicode ke base pr string ka compare hota hai
print(ord("A"))
print(ord("B"))

print("GHB">"ABC") # yaha bhi unicode ke base pr compare hoga, 1st character ka 1st character se , 2nd ka 2nd se....
```

```python
# Logical Operator------------------------------------
           # Three types - 1) and operator -  All condition must be true then return true until return false
                  #        2) or operator - anyone condition must bhi true then return true until return false
                       #   3) not operator - Reverse the boolean value

print(12>6 and 123>78 and 34==34)
print(12>6 and 123>78 and 34==34 and 56!=56)
print(78>56 or 45==67 or 43>78)
print(12==12)
print(not 12==12)  #reverse the boolean value

# PRACTICE SERIES-----------------

print(126>130)
print((456==456)!=(235==236))  # print( True != False) 
print(12<10 or 45==56 or 69>70 or 15!=13)
print(True and bool(0))
```
