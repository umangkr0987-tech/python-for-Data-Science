```python
# Create a random no. guessing game-----------------------( Use while loop)

import random  # "random" python ka in-built library hai, jisse random value generate ho sakta hai, isiliye isko pahle import krenge

nbe=random.randint(1,10)   # .ranint(a,b) => iske a aur b ke bich koe bhi random integer generate ho sakta hai
print(nbe)
tries=0
while True:
    guess=int(input("Enter the number between 1 to 10:"))

    if guess==nbe:
        tries+=1
        print(f"You are right in {tries} tries !")
        break

    elif guess<nbe:
        tries+=1
        print("You are wrong! choose no. little higher...")

    elif guess>nbe:
        tries+=1
        print("You are wrong! choose no. little lower...")   
```
