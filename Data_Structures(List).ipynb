```python
# Data Structures--- jb bahut sara data/value ko single variable me store krna padta hai , to Data Structures ka use hota hai..

# Pyhton me in-built 4 types of Data Structures hota hai => List,Tuple, Dictionary , Set

#List Data Structure-----------------------------------------------------------------------------
                         # Syntax: variable_name= [value1,value2,value3,......]

#Property of List--------------------- 1) Mutable :List create krne ke baad uska value ko change kiya ja sakta hai..
                                    #  2) Duplicate: List duplicate value allow krta hai
                                    #  3) Ordered : List me value indexed wise store rhte hia , jisse index ke through value ko access kiya ja sakta hai..
                                    #  4) Hetrogenous : List me multiple datatype ka value , function etc. sab ek hi List me store ho sakta hai...

a=[54,89,3,54,98.8,True,print()]  # List create
print(a)                           # Print list
 
a[1]=3       # List ke index no. 1 wala element ka value change kr diye (Mutable)
print(a)
print(type(a))

# List me "Indexing" aur "Slice" same hota hai string jaisa....

b=[67,54,34,32.12,45,98,56.8]

print(b[2])   # Index ke through access value
print(b[-1])  # Negative indexing ke help se bhi access kr sakte hai
print(b[1:5:1])  # Slice krna ( list ka koe part access krna)

# List me sara value pr "Traversing" bhi same string traversing jaisa hi "Loop" ke help se hota hai...
c= [12,23,34,45,56,67,78]

#Using index Traversing-----------
for i in range(len(c)):
    print(c[i])

# Using direct values traversing------
for i in c:
    print(i)

#List Method--------------------------------------------------------------------
                  #Method ko access krne ka Syntax---- List_variable_name.method_name()
    
# 1) append(value)  => ye value ko List ke last me add kr deta hai
l= [2,4,6,7,9,4,7]
l.append(17)    
l.append(True)
print(l)

# 2) insert(index_no,value) => ye value ko given index no. ke hissab se add kr deta hai
m=[1,2,3,4,5]
m.insert(2,True)
print(m)

# 3) remove(value) => ye jo value pass hota hai, uss value ko List se remove kr deta hai, lakin only first occurance (jo first baar List me aaya ho...)
n=[2,3,4,5,6,7,4,5]
n.remove(4)           # jo first 4 hoga ,wahi sirf remove hoga, sara 4 ni
print(n)

# 4) Count(value) => ye jo value pass hota hai, uo kitna baar aaya hai List me ,uo count kr deta hai, jisko humlog ek variable me store kr sakte hai
z=[4,5,6,7,8,4,5,4,5,4,5]
coun= z.count(5)
print(coun)

# 5) sort() => ye List ke value ko ascending order me sort kr deta hai....
x=[0,3,4,5,7,3,8,True]
x.sort()
print(x)            # Sorting me Boolean value ko integer me convert krke (True=1,False=0) ,fir sort kiya jata hai

# 6) reverse() => ye List ke value ko reverse kr deta hai
h=[65,98,43,65,32,9]
h.reverse()
print(h)

# 7) copy() => isse List ka duplicate copy bana sakte hai, jisko ek variable me store kiya ja sakta hai aur fir use kr sakte hai
f=[65,23,45,67,89,33]
Pee=f.copy()      # List ka duplicate copy ban gya jisko "Pee" variable me store kr liye
print(Pee)

# 8) clear() => isse list ke sara elment ko remove kr sakte hai
r=[4,5,6,7,8,9,23]
r.clear()
print(r)

# 9) index(value) => isse jo value pass hoga , uss value ka List me index no. pta kr sakte hai,lakin only first occurance (jo first baar List me aaya ho...), jisko ek varible me store krkr baad me use kr sakte hai
w=[5,2,2,3,8,9]
pos=w.index(2)     # 2 ka index no. List me pta find krna
print(pos)

# 10) pop(index_no.) => isse index no. wise element ko List se remove krke , uss element ko ek variable me store kr sakte hai
s=[4,6,3,8,2,9]
st=s.pop(3)       #index no.= 3 wala element ko List se remove kr, usko "st" variable me store krna
print(st)
print(s)

#Practice Series--------------------------------------------------------------------------------------

# Print positive and negative element on the List
um=[-98,76,-54,87,-43,93,6,-2]
bn=[]
cn=[]
for i in range(len(um)):
    if um[i]<0:
        bn.append(um[i])
    else:
        cn.append(um[i])

print(f"Negative element is :{bn}")  
print(f"Positive element is :{cn}")  

# Mean of the List----------------------------------------------------------------------------
y=[1,2,3]
add=0;
count=0
for i in y:
    add+=i
    count+=1

print(f"Mean of the List is: {add/count}")
print(f"Mean of the List is: {add/len(y)}")   # Total count = length of List

# Find the greatest element and print its index too.
e=[54,65,98,34,78,23,43,89]
gt=0
index=0

for i in range(len(e)):
    if gt<e[i]:
        gt=e[i]
        index=i

print(f"Greatest no. is: {gt} whose index is: {index}")   

# Find second Largest no. 
f=[54,76,32,87,53,94,65,92,99]
gt=0
s_gt=0

for i in range(len(f)):
    if gt<f[i]:
       s_gt=gt
       gt=f[i]
    elif f[i]>s_gt:
        s_gt=f[i]
        
print(f"Second largest no. is: {s_gt}")

# Check the List is sorted or not?
v=[4,5,6,9,7,8,9]

for i in range(len(v)-1):
    if v[i]<v[i+1]:
        continue
    else:
        print("The List is not sorted")
        break
else:
    print("The list is sorted")
   
# Remove duplicate element in List
li=[4,6,8,9,3,6,4,9,8,2,1,5,7,8]
re=[]
for i in li:
   if i not in re:   # "not in" operator => ye check krta hai ki koe element kisi List/tuple/string... me present ni hai ( agar element present ni rhta hai, to  condition "True" ho jata hai)
       re.append(i)

print(re)            
```
